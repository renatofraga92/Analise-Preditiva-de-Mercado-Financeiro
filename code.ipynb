```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Definir o ticker da ao (ex: Apple)
ticker_symbol = "AAPL"

# Obter dados histricos
# period pode ser "1d", "5d", "1mo", "3mo", "6mo", "1y", "2y", "5y", "10y", "ytd", "max"
# interval pode ser "1m", "2m", "5m", "15m", "30m", "60m", "90m", "1h", "1d", "5d", "1wk", "1mo", "3mo"
# (Note: '1m' interval only for last 7 days, others have limitations)
data = yf.download(ticker_symbol, period="1y", interval="1d")

print(data.head())

# Plotar o preo de fechamento
plt.figure(figsize=(12, 6))
plt.plot(data['Close'])
plt.title(f'Preo de Fechamento de {ticker_symbol}')
plt.xlabel('Data')
plt.ylabel('Preo de Fechamento')
plt.grid(True)
plt.show()

# Calcular retornos dirios logartmicos
data['Log_Returns'] = np.log(data['Close'] / data['Close'].shift(1))

# Plotar retornos
plt.figure(figsize=(12, 6))
plt.plot(data['Log_Returns'])
plt.title(f'Retornos Logartmicos Dirios de {ticker_symbol}')
plt.xlabel('Data')
plt.ylabel('Retorno Logartmico')
plt.grid(True)
plt.show()
```
